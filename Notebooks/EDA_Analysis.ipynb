```python
import pandas as pd
import plotly.express as px
```

```python
nav_df = pd.read_csv("../Data/Processed/nav_history_cleaned.csv")

nav_df["date"] = pd.to_datetime(nav_df["date"])

daily_nav = nav_df.groupby("date")["nav"].mean().reset_index()

fig = px.line(daily_nav, x="date", y="nav", title="Daily NAV Trend (2022-2026)") 
fig.show()
```

```python
df = pd.read_csv("../Data/Processed/fund_master_cleaned.csv")

print(df.head())
print(df.shape)
print(df.columns)
```

```python
fund_house = df["fund_house"].value_counts().reset_index()
fund_house.columns = ["Fund House", "Count"]

fig = px.bar(
    fund_house,
    x = "Fund House",
    y = "Count",
    title = "Number of Schemes by Fund Mouse"
)
fig.show()
```

```python
import pandas as pd
import plotly.express as px

df= pd.read_csv("../Data/Processed/aum_by_fund_house_cleaned.csv")


print(df.columns)
print(df.head())
```

```python
import plotly.express as px
df["date"] = pd.to_datetime(df["date"])
df["year"] = df["date"].dt.year.astype(str)

fig = px.bar(
    df,
    x="fund_house",
    y="aum_lakh_crore",
    color="year",
    barmode="group",
    title="AUM Growth by Fund House (2022-2025)"
)
fig.update_layout(
    xaxis_title="Fund House",
    yaxis_title="AUM (Lakh Crore)",
    xaxis_tickangle=-45
)
fig.show()
```

```python
import pandas as pd

df = pd.read_csv("../Data/Processed/category_inflowa_cleaned.csv")

print(df.columns)
print(df.head())
```

```python
import pandas as pd

df = pd.read_csv("../Data/Processed/investor_transactions_cleaned.csv")

print(df.columns)
print(df.head())
```

```python
import pandas as pd
 
df= pd.read_csv("../Data/Processed/Investor_transactions_cleaned.csv")

print(df.columns)
```

```python
import plotly.express as px

state_data = df.groupby("state")["amount_inr"].sum().reset_index()

fig = px.bar(
    state_data,
    x="state",
    y="amount_inr",
    title="SIP_Amount_by State"

)
fig.show()
```

```python
import plotly.express  as px 

tier_data = df["city_tier"].value_counts().reset_index()
tier_data.columns = ["City Tier", "Count"]

fig = px.pie(
    tier_data,
    names="City Tier",
    values="Count",
    title="T30 vs B30 City Tier Distribution"
)
fig.show()
```

## Key EDA Findings

1. SBI Mutual Fund has the highest Assets Under Management (AUM) among all fund houses.
2. Daily NAV shows an overall upward trend with normal market fluctuations.
3. SIP inflows have increased steadily over the years, indicating growing investor participation.
4. Large Cap, Mid Cap and Small Cap categories receive the highest inflows.
5. Most investors belong to the 26–35 and 36–45 age groups.
6. SIP transactions are more frequent than Lumpsum and Redemption transactions.
7. T30 cities contribute a larger share of investments compared to B30 cities.
8. Equity schemes dominate the mutual fund industry across categories.
9. Monthly SIP AUM has shown consistent long-term growth.
10. Investor participation has increased significantly in recent years.
